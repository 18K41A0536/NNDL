```python
it=0
max_it=400
n=0.01
x=2
y=3

deltax=1
deltay=1
def der(x):
  x_der=(2*x)
  return x_der
def der(y):
  y_der=(2*y)
  return y_der
while it<max_it:
  x1=x
  y1=y
  deltax=-n*der(x1)
  deltay=-n*der(y1)
  x=x+deltax
  y=y+deltay
  it=it+1
  print("total iteration",it);
  print("x value is",x)
  print("y value is",y)
print("local minimum at value",x)
print("local minimum value for y",y)  

```
