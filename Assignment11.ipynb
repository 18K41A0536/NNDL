```python
import pandas as pd
import numpy as np
```

```python
data={'x':[0.2,0.4,0.6,0.8],
      'y':[3.4,3.6,4.2,4.6]
      }
df=pd.DataFrame(data)     
```

```python
df
```

```python
max_itr=100
itr=0
n=0.1
r=0.9
vm=vc=0
ns=4
m=1
c=-1

while itr<max_itr:
  s=0
  i=0
  while s<ns:
    for i in range(len(df)):
      m1=m
      c1=c
      vm1=vm
      vc1=vc
      one=(df.iloc[i]['y'])
      two=(df.iloc[i]['x'])
      gm=-((one-(m1+r*vm1)*two-(c1+r*vc))*two)
      gc=-(one-(m1+r*vm1)*two-(c1+r*vc))
      vm=r*vm1-n*gm
      vc=r*vc1-n*gc
      m=m+vm
      c=c+vc
      s=s+1
      print("m is",m)
      print("c is",c)
  itr=itr+1
  print("itr",itr)
print("m value",m)
print("c value",c)  

```
